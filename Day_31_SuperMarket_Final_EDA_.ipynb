```python
category_sales = df.groupby('Category')['Sales'].sum()
category_sales
```

```python
Sub_Caterory = df.groupby('Sub-Category')['Sales'].sum()
Sub_Caterory
```

```python
sns.barplot(x=Sub_Caterory.values, y=Sub_Caterory.index, palette = 'rocket')
plt.title('Total Sales by Sub-Category')
plt.xlabel('Sub-Category')
plt.ylabel('Total Sales')
```

```python
Sub_Caterory = df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False).head(5)
Sub_Caterory
```

```python
sns.barplot(x=Sub_Caterory.index, y=Sub_Caterory.values, palette = 'rocket')
plt.title('Total Sales by Sub-Category')
plt.xlabel('Sub-Category')
plt.ylabel('Total Sales')
plt.legend(title='Category',labels=Sub_Caterory.index, loc='upper right')
```

```python
city = df.groupby('City')['Sales'].sum().sort_values(ascending=False).head(10)
city
```

```python
sns.barplot(x=city.values, y=city.index, palette = 'rocket')
plt.title('Total Sales by City')
plt.xlabel('City')
plt.ylabel('Total Sales')
```
